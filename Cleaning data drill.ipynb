# Central tendencies

```python
df2['COST (£) charged to Wellcome (inc VAT when charged)'] = pd.to_numeric(df2['COST (£) charged to Wellcome (inc VAT when charged)'])
df3_mean = df2.groupby('Journal title', as_index=False).sum()
df3_mean = df3_mean.sort_values('COST (£) charged to Wellcome (inc VAT when charged)', ascending = False)
df3_mean.head(10)
```

```python
df3_mean = df2.groupby('Journal title', as_index=False).mean()
df3_mean = df3_mean.sort_values('COST (£) charged to Wellcome (inc VAT when charged)', ascending = False)
df3_mean.head(10)
```

```python
df3_mean = df2.groupby('Journal title', as_index=False).median()
df3_mean = df3_mean.sort_values('COST (£) charged to Wellcome (inc VAT when charged)', ascending = False)
df3_mean.head(10)
```
